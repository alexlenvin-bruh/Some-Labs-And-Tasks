# Нормальное распределение вероятостей. Метод полярных координат.

## Метод полярных координат

```python
import numpy as np

#>>>-----------------Для получения равномерно распределённых чисел---------------------->
print("Ведите числа:")
print("1) k для X0 = 2 + 4 * k")
print("2) n для кол-ва чисел")
print("3) e для Xn+1 = Xn * (Xn + 1) mod 2^e")

A = [int(i) for i in input().split()]  # k N e

print(" >>>>>----------------------------------------------------->")
print("k =", A[0], "n =", A[1],  "e =", A[2])

Xn = 4 * A[0] + 2
m = 2 ** A[2]
seq = list()            #Сюда пишем последовательность

for i in range(0, A[1]):
    #>>>-----------------Получаем х и у из равномерно распределённых чисел---------------------->
    Xn = (Xn * (Xn + 1)) % m    #for 1st number
    x = 2 * Xn / m - 1

    Xn = (Xn * (Xn + 1)) % m    #for 2nd number
    y = 2 * Xn / m - 1

    #>>>-----------------Проверяем пригодность получен-ных х, у---------------------->
    S = x ** 2 + y ** 2
    if S >= 1 or S == 0:
        continue                #если «расстояние» не удовлетворяет нас, то выбираем другие два числа

    #>>>-----------------Получаем и записываем элементы в итоговую последовательность---------------------->
    num = x * (- 2 * np.log(S) / S) ** (1/2)
    seq.append(num)
    num = y * (- 2 * np.log(S) / S) ** (1/2)
    seq.append(num)

print(" ")
print(seq)
print(len(seq))
```

```python
# Проверяем гипотезу, что данные из N(0, 1)
from scipy import stats

ks_stat, p_value = stats.kstest(seq, 'norm', args=(0, 1))

print(f"KS статистика: {ks_stat:.4f}")
print(f"p-value: {p_value:.4f}")
print("Результат:", "✓" if p_value > 0.05 else "✗")
```

```python
import math

def norm_cdf(x):
    """CDF стандартного нормального распределения N(0,1)"""
    if x < 0:
        return 1 - norm_cdf(-x)
    
    t = 1 / (1 + 0.2316419 * x)
    d = 0.3989423 * math.exp(-x * x / 2)
    prob = d * t * (0.3193815 + t * (-0.3565638 + t * (1.781478 + t * (-1.821256 + t * 1.330274))))
    return 1 - prob

# КС-тест
seq_sorted = sorted(seq)
n = len(seq_sorted)
max_diff = 0

for i, x in enumerate(seq_sorted):
    # Эмпирическая CDF в точке x
    F_emp = (i + 1) / n
    
    # Теоретическая CDF в точке x
    F_teor = norm_cdf(x)
    
    # Разница
    diff = abs(F_emp - F_teor)
    max_diff = max(max_diff, diff)

# Критическое значение для α = 0.05
D_crit = 1.36 / (n ** 0.5)

print(f"КС-статистика D = {max_diff:.4f}")
print(f"D_критическое (α=0.05) = {D_crit:.4f}")
print("Результат:", "✓" if max_diff < D_crit else "✗")
```
